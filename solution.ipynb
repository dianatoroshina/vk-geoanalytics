```python
import pandas as pd
from joblib import load
```

Загрузим уже обученную модель

```python
model = load('model_geo.joblib')
test_df = pd.read_csv('test_modified.csv')
```

```python
test_df['score'] = model.predict(test_df.drop(labels=['id'], axis=1))
print(test_df)
```

Сохраним результаты в виде, соответствующему sample_submission.csv

```python
test_sub = pd.DataFrame(test_df, columns=['id', 'score'])
print(test_sub)
```

```python
test_sub.to_csv('submission.csv')
```
